```python
%sql
USE CATALOG students_data;
USE SCHEMA `team3-taxi`;
```

```python
%sql
    
CREATE OR REPLACE TABLE fact_trip_silver AS
SELECT
    -- Surrogate technical key for the trip
    MD5(
        CONCAT(
            VendorID, '|',
            tpep_pickup_datetime, '|',
            tpep_dropoff_datetime, '|',
            pickup_longitude, '|',
            pickup_latitude, '|',
            dropoff_longitude, '|',
            dropoff_latitude
        )
    ) AS trip_id,

    -- Natural keys
    VendorID              AS vendor_id,
    RateCodeID            AS rate_code_id,
    payment_type          AS payment_type_id,

    -- Timestamps
    tpep_pickup_datetime  AS pickup_datetime,
    tpep_dropoff_datetime AS dropoff_datetime,

    -- Derived duration (minutes)
    DATEDIFF(MINUTE, tpep_pickup_datetime, tpep_dropoff_datetime) AS trip_duration_minutes,

    -- Locations
    pickup_latitude,
    pickup_longitude,
    dropoff_latitude,
    dropoff_longitude,

    -- Measures
    passenger_count,
    trip_distance,
    fare_amount,
    extra,
    mta_tax,
    tip_amount,
    tolls_amount,
    improvement_surcharge,
    total_amount,

    -- Attributes
    store_and_fwd_flag

FROM stg_journey;
```
